```python
import pandas as pd
import matplotlib.pyplot as plt
import re
from matplotlib.ticker import FuncFormatter

# Fetch all the tables present in the HTML content
url = 'https://24365withblinks.com/about_records'
all_tables = pd.read_html(url)

# Assume you want to create a bar chart from the first table
if all_tables:
    first_table = all_tables[0]

    # Assuming the '24 Hour Views' column exists in the table, adjust as needed
    if '24 Hour Views' in first_table.columns:
        # Extract only the numeric part from the '24 Hour Views' column
        first_table['24 Hour Views'] = first_table['24 Hour Views'].apply(
            lambda x: int(re.search(r'\d+', x.replace('.', '').replace(',', '')).group()) if pd.notnull(x) else 0
        )

        # Sort the data based on '24 Hour Views' column for better visualization
        first_table = first_table.sort_values(by='24 Hour Views', ascending=False)

        # Define a function to format y-axis ticks as millions
        def millions_formatter(x, pos):
            return f'{x / 1e6:.1f}M'

        # Plot the bar chart
        plt.figure(figsize=(12, 8))
        plt.bar(first_table['Song'], first_table['24 Hour Views'], color='skyblue')
        plt.xlabel('Song')
        plt.ylabel('24 Hour Views')
        plt.title('Bar Chart of Views in 24 Hrs')
        plt.xticks(rotation=45, ha='right')
        plt.gca().get_yaxis().set_major_formatter(FuncFormatter(millions_formatter))  # Apply the formatter
        plt.tight_layout()

        # Display the plot
        plt.show()
    else:
        print("The specified column '24 Hour Views' does not exist in the table.")
else:
    print("No tables found on the provided URL.")
```
